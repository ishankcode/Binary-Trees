Tree Traversal:

1. Preorder  - root->root.left->root.right
2. Postorder - root.left->root.right->root
3. inorder   - root.left->root->root.right
4. levelorder_traversal- level wise we traverse through a tree


```python
class binarytreenode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
def takeinput():
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=binarytreenode(rootdata)
    lefttree=takeinput()
    righttree=takeinput()
    root.left=lefttree
    root.right=righttree
    
    return root

def print_pre_order(root):
    if root is None:
        return
    
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    print_pre_order(root.left)
    print_pre_order(root.right)

def print_post_order(root):
    if root is None:
        return
    print_post_order(root.left)
    print_post_order(root.right)
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()

def print_inorder(root):
    if root is None:
        return
    print_inorder(root.left)
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    print_inorder(root.right)

    
c=takeinput()
print('preorder')
print_pre_order(c)
print('postorder')
print_post_order(c)
print('inorder')
print_inorder(c)
```

```python
class binarytreenode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
        
def nodes_largest_data(root):
    # if root is empty then i have no data so should return huge negative no. like -infinity 
    # or if all data is positive then -1
    if root  is None:
        return -1
    a=nodes_largest_data(root.left)
    b=nodes_largest_data(root.right)
    return max(root.data,a,b)

def nodes_greater_than_x(root,x):
    if root  is None:
        return 0
    
    a=nodes_greater_than_x(root.left,x)
    b=nodes_greater_than_x(root.right,x)
    if root.data>x:
        return 1+a+b
    else:
        return a+b

# if single node present height =1

def height_of_tree(root):
    if root is None:
        return 0
    a=height_of_tree(root.left)
    b=height_of_tree(root.right)
    return 1+max(a,b)

def no_of_leaf_nodes(root):
    if root is None:
        return 0
    
    if root.left is None and root.right is None:
        return 1
    
    a=no_of_leaf_nodes(root.left)
    b=no_of_leaf_nodes(root.right)
    
    return a+b

def print_at_depth(root,k):
    if root  is None:
        return 
    if k==0:
        print(root.data)
        return 
    print_at_depth(root.left,k-1)
    print_at_depth(root.right,k-1)

def print_at_depth_m2(root,k,d=0):
    if root  is None:
        return 
    if k==d:
        print(root.data)
        return 
    print_at_depth_m2(root.left,k,d+1)
    print_at_depth_m2(root.right,k,d+1)
    
def replace_node_with_depth(root,d=0):
    if root is None:
        return 
    print(d)
    replace_node_with_depth(root.left,d+1)
    replace_node_with_depth(root.right,d+1)
    
def takeinput():
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=binarytreenode(rootdata)
    lefttree=takeinput()
    righttree=takeinput()
    root.left=lefttree
    root.right=righttree
    
    return root

def printdetailed(root):
    if root is None:
        return
    
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    printdetailed(root.left)
    printdetailed(root.right)

#tree is in level(wwise): 1 2 3 4 10 5 8 -1 -1 12 15 -1 -1 -1 -1
# no. of nodes greater than 9
# at depth 3

c=takeinput()
printdetailed(c)
print('Largest',nodes_largest_data(c))
no=int(input())
print('no. of nodes larger than',no,'are :',nodes_greater_than_x(c,no))
print('height of tree is :',height_of_tree(c))
print('no of leeaf nodes are',no_of_leaf_nodes(c))
k=int(input())
print('at given depth',print_at_depth(c,k))
print('at given depth method 2',print_at_depth_m2(c,k))
print('preorder replacement with height')
replace_node_with_depth(c)
```

Check if  the binary tree is  balanced or not:

For each ode inn a tree, that nodes left and right subtree should not have height diffrence of more than 1

```python
class binarytreenode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

def height(root):
    if root is None:
        return 0
    a=height(root.left)
    b=height(root.right)
    
    return 1+max(a,b)
    
def check_balanced(root):
    if root is None:
        return True
    leftheight=height(root.left)
    rightheight=height(root.right)
    
    if leftheight-rightheight>1 or rightheight-leftheight>1:
        return False
    
    l=check_balanced(root.left)
    r=check_balanced(root.right)
    
    if l is True and r is True:
        return True
    else:
        return False
# here we are calculating height at each node
# then again we are travelling each node to check if they are overbalanced or  not 

def check_balanced2(root):
    if root is None:
        return 0,True
    lh,left_balanced=check_balanced2(root.left)
    rh,right_balanced=check_balanced2(root.right)
    h=1+max(lh,rh)
    if lh-rh>1 or rh-lh>1:
        return h,False
    if left_balanced and right_balanced :
        return h,True
    else:
        return h,False
    
def takeinput():
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=binarytreenode(rootdata)
    lefttree=takeinput()
    righttree=takeinput()
    root.left=lefttree
    root.right=righttree
    
    return root

def printdetailed(root):
    if root is None:
        return
    
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    printdetailed(root.left)
    printdetailed(root.right)

c=takeinput()
printdetailed(c)
print(check_balanced(c))
a,b=check_balanced2(c)
print(b)

```

Diameeter of binary tree:  distance btween 2 farthest nodes

```python
class binarytreenode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

def height(root):
    if root is None:
        return 0
    a=height(root.left)
    b=height(root.right)
    
    return 1+max(a,b)
    
def diameter(root):
    dia,depth=0,0
    if root is None:
        return dia,depth
    
    leftdia,ldepth=diameter(root.left)
    rightdia,rdepth=diameter(root.right)
    
    depth=1+max(ldepth,rdepth)
    dia=ldepth+rdepth
    
    return max(leftdia,rightdia,dia),depth
    
def takeinput():
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=binarytreenode(rootdata)
    lefttree=takeinput()
    righttree=takeinput()
    root.left=lefttree
    root.right=righttree
    
    return root

def printdetailed(root):
    if root is None:
        return
    
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    printdetailed(root.left)
    printdetailed(root.right)

c=takeinput()
printdetailed(c)
print(diameter(c))
```

```python
# Cant use recursion so use itteratively and it's so similar to FIFO so we use queue
# take input and add to queue,  whle queue is not take it out and ask its  children if valid then add them to the queue
class binarytree:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
import queue
def input_levelwise():
    q=queue.Queue()
    print('enter root')
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=binarytree(rootdata)
    q.put(root)
    
    while (not(q.empty())):
        current=q.get()
        print('enter left child of',current.data)
        leftdata=int(input())
        if leftdata!=-1:
            left_child=binarytree(leftdata)
            current.left=left_child
            q.put(left_child)
        print('enter right child of',current.data)
        rightdata=int(input())
        if rightdata!=-1:
            right_child=binarytree(rightdata)
            current.right=right_child
            q.put(right_child)
            
    return root

def printdetailed(root):
    if root is None:
        return
    
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data,end=',')
    if root.right  is not None:
        print('R',root.right.data,end='')
    print()
    printdetailed(root.left)
    printdetailed(root.right)



c=input_levelwise()
printdetailed(c)
```

Construct tree using inorder and preeorder

```python
import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def buildTreePreOrder(preorder, inorder):
    #example
    # preOrder: 1 2 3 4 15 5 6 7 8 10 9 12
    # inOrder: 4 15 3 2 5 1 6 10 8 7 9 12

    if len(preorder)==0:
        return None
    # step 1 : find root  1st which can easily be found from preorder
    rootdata=preorder[0]
    root=BinaryTreeNode(rootdata)
    # step 2 : search for the root  in inorder
    for i in range(0,len(inorder)):
        if inorder[i]==rootdata:
            root_index=i
            break
    # everything left to found index in inorder is left subtree
    # everything to right is rightsub tree
    left_in=inorder[0:root_index]
    right_in=inorder[root_index+1:]
    
    # now we are finding left subtree elements in preorder
    left_pre=preorder[1:len(left_in)+1]
    right_pre=preorder[len(left_in)+1:]
    # now we call recursion again giving this as input   
    leftsub=buildTreePreOrder(left_pre,left_in)
    rightsub=buildTreePreOrder(right_pre,right_in)
    
    #now we connect the 2 subtrees
    root.left=leftsub
    root.right=rightsub
    
    return root

def printLevelATNewLine(root):
    # Given a binary tree, print the level order traversal. Make sure each level
    # start in new line.
    if root==None:
        return
    inputQ = queue.Queue()
    outputQ = queue.Queue()
    inputQ.put(root)
    while not inputQ.empty():
        while not inputQ.empty():
            curr = inputQ.get()
            print(curr.data, end=' ')
            if curr.left!=None:
                outputQ.put(curr.left)
            if curr.right!=None:
                outputQ.put(curr.right)
        print()
        inputQ, outputQ = outputQ, inputQ

# Main
n=int(input())
preorder = [int(i) for i in input().strip().split()]
inorder = [int(i) for i in input().strip().split()]
root = buildTreePreOrder(preorder, inorder)
printLevelATNewLine(root)
```

construct tree using postorder and inorder

```python
import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def buildTreePostOrder(postorder, inorder):
    # Given Postorder and Inorder traversal of a binary tree, create the binary
    # tree associated with the traversals.You just need to construct the tree
    # and return the root. For 8 Nodes with following input:
    # postOrder: 8 4 5 2 6 7 3 1
    # inOrder: 4 8 2 5 1 6 3 7
    if len(postorder)==0:
        return None
    
    rootdata=postorder[-1]
    root=BinaryTreeNode(rootdata)
    for i in range(0,len(inorder)):
        if inorder[i]==rootdata:
            root_index=i
            break
            
    left_in=inorder[0:root_index]
    right_in=inorder[root_index+1:]
    
    left_post=postorder[0:len(left_in)]
    right_post=postorder[len(left_in):-1]
    
    leftsub=buildTreePostOrder(left_post,left_in)
    rightsub=buildTreePostOrder(right_post,right_in)
    
    root.left=leftsub
    root.right=rightsub
    
    return root

def printLevelATNewLine(root):
    # Given a binary tree, print the level order traversal. Make sure each level
    # start in new line.
    if root==None:
        return
    inputQ = queue.Queue()
    outputQ = queue.Queue()
    inputQ.put(root)
    while not inputQ.empty():
        while not inputQ.empty():
            curr = inputQ.get()
            print(curr.data, end=' ')
            if curr.left!=None:
                outputQ.put(curr.left)
            if curr.right!=None:
                outputQ.put(curr.right)
        print()
        inputQ, outputQ = outputQ, inputQ

# Main
n=int(input())
postOrder = [int(i) for i in input().strip().split()]
inorder = [int(i) for i in input().strip().split()]
root = buildTreePostOrder(postOrder, inorder)
printLevelATNewLine(root)
```

Insert Duplicate Nodes:

```python
import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insertDuplicateNode(root):
    # Given a BST with N number of nodes, for each node create a new duplicate
    # node, and insert that duplicate as left child of the original node.
    if root is None:
        return
    # save left data in left
    left=root.left
    # Now  assign root.left to duplicate of root
    root.left=BinaryTreeNode(root.data)
    # now this left left will point to original left
    root.left.left=left
    
    #now call recursion on both left and right tree
    
    insertDuplicateNode(root.left.left)
    insertDuplicateNode(root.right)
    

def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

def printLevelATNewLine(root):
    # Given a binary tree, print the level order traversal. Make sure each level
    # start in new line.
    if root==None:
        return
    inputQ = queue.Queue()
    outputQ = queue.Queue()
    inputQ.put(root)
    while not inputQ.empty():
        while not inputQ.empty():
            curr = inputQ.get()
            print(curr.data, end=' ')
            if curr.left!=None:
                outputQ.put(curr.left)
            if curr.right!=None:
                outputQ.put(curr.right)
        print()
        inputQ, outputQ = outputQ, inputQ

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
insertDuplicateNode(root)
printLevelATNewLine(root)
## Example output

#1
#2 3
##

#1
#1 3
#2 3
#2
```

```python

import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
# Took hiuge nnegative and positive no. default
INT_MIN = -2147483648
INT_MAX = 2147483647
# Problem ID 1567, Find the minimum and maximum value
def minMax(root):
    # Given a binary tree, find and return the min and max data value of given
    # binary tree. Return maximum and minimum from this function only. Largest 
    # and smallest data possible is provided through INT_MAX and INT_MIN
    
    # if  root is none theen return these 2 values
    if root is None:
        return INT_MAX,INT_MIN 
    
    # store root's data 
    x=root.data
    # and call recursion on left and right subtree
    a,b = minMax(root.left)
    c,d = minMax(root.right)
    
    return min(a,c,x), max(b,d,x)
    

def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
minimum, maximum = minMax(root)
print(maximum, minimum)
```

Print all nodes at distance k from a given node

```python
import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def print_at_depth(root,k):
    if root is None:
        return 
    if k==0:
        print(root.data)
        return
    
    print_at_depth(root.left,k-1)
    print_at_depth(root.right,k-1)
    
    
def nodes_at_k(root,node,k):
    if root is None:
        return -1
    # root is same as the given node then i am printing all nodes at depth k from this
    if root.data==node:
        print_at_depth(root,k)
        return 0  # this is the distance btw root and the given node, iw ill be returning this
    
    # now calling on left sub tree
    ld=nodes_at_k(root.left,node,k)
    if ld!=-1:
        # if distance from this to given node +1 gives root
        if ld+1==k:
            print(root.data)
        else:
            # otherwise k-(ld+1) is  still the depth remaining
            # so now as right root is even one distance apart, so k-ld-2is the level at which any node found should be printed
            print_at_depth(root.right,k-ld-2)
        
        return ld+1
    
    # same process is  repeated for right node as well
    rd=nodes_at_k(root.right,node,k)
    if rd!=-1:
        if rd+1==k:
            print(root.data)
        else:
            print_at_depth(root.left,k-rd-2)
        
        return rd+1
    
    return -1
    

    
def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
node=int(input())
k=int(input())
nodes_at_k(root,node,k)
```
